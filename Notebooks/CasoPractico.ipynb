### Actividad 2. Análisis de relación entre variables

•	Calcular la matriz de correlación de las variables numéricas.
•	Ordenar las correlaciones tomando como referencia la variable Prestige.
•	Identificar las variables con relación positiva más fuerte.
•	Identificar relaciones negativas o débiles, si existen.
•	Explicar si los resultados son coherentes desde el punto de vista social, laboral o productivo.
•	Advertir que correlación no significa causalidad.


```python
# Seleccionamos solo las columnas con tipos de datos numéricos (int64, float64).
# Esto es importante para calcular correctamente la matriz de correlación.
variables_numericas = df.select_dtypes(include=['int64', 'float64'])

# Calculamos la matriz de correlación entre todas las variables numéricas seleccionadas.
# La correlación mide la fuerza y dirección de la relación lineal entre dos variables.
matriz_correlacion = variables_numericas.corr()

# Imprimimos la matriz de correlación completa, redondeando los valores a 3 decimales para mayor claridad.
# Salida: Matriz de correlación (DataFrame).
print(matriz_correlacion.round(3))

# Ordenamos las correlaciones de la variable 'prestige' de forma descendente.
# Esto nos permite identificar rápidamente qué variables tienen la relación más fuerte (positiva y negativa) con el prestigio.
# Salida: Serie con los valores de correlación con 'prestige', ordenados.
print(matriz_correlacion['prestige'].sort_values(ascending=False).round(3))
```

#### Actividad 4. Análisis de agrupación de ocupaciones
En esta actividad se aplicará clusterización para identificar grupos de ocupaciones con características similares. El estudiante debe construir al menos dos modelos: uno con 4 grupos y otro con 5 grupos, y luego justificar cuál considera más útil para interpretar el caso.

•	Seleccionar variables numéricas relevantes para la clusterización.

•	Escalar los datos antes de aplicar K-Means.

•	Ejecutar K-Means con 4 clústeres.

•	Ejecutar K-Means con 5 clústeres.

•	Comparar los centroides de ambos modelos.

•	Interpretar qué caracteriza a cada grupo.

•	Identificar cuáles grupos tienden a tener mayor prestigio y cuáles menor prestigio.


```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Definimos las columnas numéricas que se utilizarán para el clustering.
# Estas son las variables que caracterizarán a cada ocupación para agruparlas.
columnas_cluster = ['education', 'income', 'women', 'prestige', 'type_num']
X = df[columnas_cluster].copy() # Creamos una copia del subconjunto de datos para el clustering.

# Inicializamos el escalador estándar (StandardScaler).
# Este escalador transforma los datos para que tengan una media de 0 y una desviación estándar de 1.
# Esto es crucial para K-Means, ya que se basa en distancias y las variables con rangos más grandes dominarían el cálculo.
escalador = StandardScaler()
# Ajustamos el escalador a los datos y luego los transformamos.
X_escalado = escalador.fit_transform(X)

# --- Modelo K-Means con 4 clústeres ---
# Inicializamos el modelo K-Means con 4 clústeres, una semilla aleatoria para reproducibilidad (random_state=42),
# y 10 inicializaciones diferentes para encontrar la mejor agrupación (n_init=10).
modelo_k4 = KMeans(n_clusters=4, random_state=42, n_init=10)
# Ajustamos el modelo a los datos escalados y predecimos el clúster al que pertenece cada ocupación.
# Asignamos la etiqueta del clúster a una nueva columna en el DataFrame original.
df['Cluster_4'] = modelo_k4.fit_predict(X_escalado)

# --- Modelo K-Means con 5 clústeres ---
# Realizamos el mismo proceso, pero esta vez con 5 clústeres.
modelo_k5 = KMeans(n_clusters=5, random_state=42, n_init=10)
df['Cluster_5'] = modelo_k5.fit_predict(X_escalado)

# --- Análisis de Centroides (Medias por Clúster) ---
# Calculamos la media de cada una de las variables de clustering para cada uno de los 4 clústeres.
# Esto nos permite entender las características promedio de las ocupaciones dentro de cada grupo.
# Salida: DataFrame con las medias de las variables por cada clúster (K=4).
print("Medias de las variables por cada uno de los 4 clústeres:")
print(df.groupby('Cluster_4')[columnas_cluster].mean().round(2))

# Realizamos el mismo cálculo para los 5 clústeres.
# Salida: DataFrame con las medias de las variables por cada clúster (K=5).
print("\nMedias de las variables por cada uno de los 5 clústeres:")
print(df.groupby('Cluster_5')[columnas_cluster].mean().round(2))
```

```python
# Obtenemos los centroides de los 4 clústeres del modelo K-Means. Estos están en la escala de los datos estandarizados.
centros_cluster_k4_escalados = modelo_k4.cluster_centers_

# Para interpretar los centroides en su escala original (antes de la estandarización),
# aplicamos la transformación inversa utilizando el mismo escalador que se usó para estandarizar los datos.
centros_cluster_k4_originales = escalador.inverse_transform(centros_cluster_k4_escalados)

# Convertimos los centroides transformados de vuelta a un DataFrame de pandas para una visualización más legible.
# Asignamos los nombres de las columnas originales para facilitar la interpretación.
df_centros_k4 = pd.DataFrame(centros_cluster_k4_originales, columns=columnas_cluster)

# Imprimimos los centroides de los 4 clústeres, redondeando los valores a 2 decimales.
# Salida: DataFrame que muestra las características promedio de cada clúster en la escala original de las variables.
print("Centroides de los 4 clústeres (valores originales):")
print(df_centros_k4.round(2))
```
